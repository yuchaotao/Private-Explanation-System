```python
import pandas as pd
```

```python
df = pd.read_csv('ipumscps.csv')
```

```python
df.columns
```

```python
attr_agg = 'INCTOT'
attr_gb = 'RELATE'
def check(df, attr_agg, attr_gb):
    return df.groupby(attr_gb).agg(answer = (attr_agg, 'mean'), size = (attr_agg, 'count')).sort_values(by='size', ascending=False)
```

```python
check(df, attr_agg, 'RELATE')
```

```python
check(df, attr_agg, 'MARST')
```

```python
check(df, attr_agg, 'SEX')
```
