```python
import pandas as pd
```

```python
header = [
   'parents',
   'has_nurs',
   'form',
   'children',
   'housing',
   'finance',
   'social',
   'health',
   'recommendation'
]
receommendation_mapping = {
    'not_recom': 1,
    'recommend': 2,
    'very_recom': 3,
    'priority': 4,
    'spec_prior': 5
}
df = pd.read_csv('nursery.data', names=header, index_col=False)
df['recommendation'] = df['recommendation'].apply(lambda x: receommendation_mapping[x])
```
